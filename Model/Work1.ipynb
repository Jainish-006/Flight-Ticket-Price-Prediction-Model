```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv("business.csv")
df.head()
```

```python
df.stop = df.stop.map(lambda x: '0-stop' if 'non-stop ' in x else x)
df.head()
```

```python
df.stop.unique()
```

```python
def convert_string_to_num(y):
    token  = y.split('-')
    if len(token) == 2:
        return ((token[0]))
    try:
        return int(y)
    except:
        return None
```

```python
df.stop = df.stop.apply(convert_string_to_num)
df = df[df.stop.notnull()]
df.head()
```

```python
df.stop.unique()
```

```python
df.isnull().sum()
```

```python
df1 = df
df1.head()
```

```python
def convert_time(x):
    tok1 = x.replace('h ','.')
    tok2 = tok1.replace('m','')
    return tok2
```

```python
df1.time_taken = df.time_taken.apply(convert_time)
df1.head()
```

```python
df1.ch_code.unique()
```

```python
df.dtypes
```

```python
df1['num_code'] = df['num_code'].map(str)
df1.dtypes
```

```python
df1['flight'] = df[['ch_code','num_code']].apply(lambda x:'-'.join(x), axis=1)
df1.head()
```

```python
df2 = df1.drop(['ch_code','num_code'], axis='columns')
df2.head()
```

```python
df2['class'] = 'Business'
df2.head()
```

```python
def convert_digittime(x):
    token = x.replace(':','.')
    return token
```

```python
df2.dep_time = df2.dep_time.apply(convert_digittime)
df2.head()
```

```python
df3 = df2
df3['dep_time'] = df3['dep_time'].map(float)
df3.dtypes
```

```python
def condition_tsplit(x):
    if x>=16.00 and x<20.00:
        y = str(x)
        token = str(x).replace(y,'Evening')
    elif x>=20.00 and x<24.00:
        y = str(x)
        token = str(x).replace(y,'Night')
    elif x>=0 and x<7.00:
        y = str(x)
        token = str(x).replace(y,'Early_Morning')
    elif x>=7.00 and x<12.00:
        y = str(x)
        token = str(x).replace(y,'Morning')
    elif x>=12.00 and x<16.00:
        y = str(x)
        token = str(x).replace(y,'Afternoon')
    return token
```

```python
df3.dep_time = df3.dep_time.apply(condition_tsplit)
df3.head()
```

```python
df3.head()
```

```python
df4 = df3
df4.arr_time = df4.arr_time.apply(convert_digittime)
df4.head()
```

```python
df4['arr_time'] = df4['arr_time'].map(float)
df4.dtypes
```

```python
df4.arr_time = df4.arr_time.apply(condition_tsplit)
df4.head()
```

```python
df5 = df4
```

```python
def convert_price(x):
    token = x.replace(',','')
    return token
```

```python
df5.price = df5.price.apply(convert_price)
df5.head()
```

```python
df5['price'] = df5['price'].map(int)
df5['time_taken'] = df5['time_taken'].map(float)
df5.dtypes
```

```python
def predict_price(airline,time,source,destination,class_type):
    al_index = np.where(X.columns=='airline_'+airline)[0][0]
    det_index = np.where(X.columns=='dep_time_'+time)[0][0]
    sc_index = np.where(X.columns=='source_'+source)[0][0]
    dest_index = np.where(X.columns=='destination_'+destination)[0][0]
    cl_index = np.where(X.columns==class_type)[0][0]

    x = np.zeros(len(X.columns))
    if al_index >= 0:
        x[al_index] = 1
        
    if det_index >= 0:
        x[det_index] = 1
        
    if sc_index >= 0:
        x[sc_index] = 1
    
    if dest_index >= 0:
        x[dest_index] = 1
            
    if cl_index >= 0:
        x[cl_index] = 1
        
    return model1.predict([x])[0]
```
